```python
chi['Name'] = chi['Name'].str.title()
chi['Position Title'] = chi['Position Title'].str.title()
chi
```

```python
chi.dropna(how='all',inplace=True)
#you can do this while importing as well
chi
```

```python
chi['Department']=chi['Department'].str.replace('MGMNT',"Management")
chi['Department'] = chi['Department'].str.title()
chi
```

```python
chi['Employee Annual Salary']=chi['Employee Annual Salary'].str.replace('$',"").astype(float)
```

```python
chi
```

```python
mask = chi['Position Title'].str.lower().str.contains('water')#startwith,endwith
chi[mask]
```

```python
chi['Name'] = chi['Name'].str.lstrip().str.rstrip()
chi['Position Title'] = chi['Position Title'].str.lstrip().str.rstrip()
chi
```

```python
chi2 = pd.read_csv('chicago.csv',index_col='Name').dropna(how='all')
chi2
```

```python
f = chi['Name'].str.split(',').str.get(0)
m = chi['Name'].str.split(',').str.get(1)
l = chi['Name'].str.split(' ').str.get(-1)
chi.insert(1,column='F Name',value=f)
chi.insert(2,column='M Name',value=m)
chi.insert(3,column='L Name',value=l)
chi.drop('Name',axis=1,inplace=True)
chi
```

```python
food = pd.read_csv('foods.csv')
food
```

```python
food.pivot_table(index='Gender',values='Spend',aggfunc=sum)
```
