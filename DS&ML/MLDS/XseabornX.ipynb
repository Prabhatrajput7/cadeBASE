```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head(5)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)#kde=line [kernal density estimation]
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')#kind customize
```

```python
sns.pairplot(tips,hue='sex')#hue category cols likie male and female
```

```python
sns.rugplot(tips['total_bill'])
```
