# which product are more sold together?

```python
dup = allmon[allmon['Order ID'].duplicated(keep=False)]
dup['Grouped'] = dup.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
dup.drop_duplicates(subset=['Order ID','Grouped'],inplace=True)
```
