```python
exn = pd.read_csv('https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD')
exn
```

```python
girls = exn[exn['Gender']=='FEMALE']
boys = exn[exn['Gender']=='MALE']
```

```python
#excel constructor
excelf = pd.ExcelWriter('babynames.xlsx')
girls.to_excel(excelf,sheet_name='Girls',index=False)
boys.to_excel(excelf,sheet_name='Boys',index=False,columns=['Year of Birth','Gender','Child\'s First Name'])
```

```python
excelf.save()
```
